```python
import pandas as pd
```

```python
cols = ['ROUTE NAME',
'ROUTE TYPE',
'TOTAL ROUTE LENGTH',
'NO OF CULVERT',
'LENGTH (IN Mtr) OF CULVERT',
'NO OF CANAL',
'LENGTH (IN Mtr) OF CANAL',
'NO OF BRIDGE',
'LENGTH (IN Mtr) OF BRIDE',
'LENGTH (IN Mtr) OF PCC',
'LENGTH (IN Mtr) OF DWC',
'LENGTH (IN Mtr) OF GI',
'LENGTH (IN Mtr) OF DWC+PCC',
'LENGTH (IN Mtr) OF ANCORING']

protection_details = pd.DataFrame(columns = cols)
protection_details
```

```python
porsa_survey = pd.read_excel('porsa_block_survey_data.xlsx', sheet_name='Sheet1')
span_ = porsa_survey[['from_gp_na', 'to_gp_name', 'span_name', 'ring_no', 'scope', 'span_id']].drop_duplicates()
span_dis= porsa_survey.groupby('span_name').agg({'distance': 'sum'})
merged_df = pd.merge(span_, span_dis, on=['span_name'], how='inner')
merged_df
```

```python
porsa_details = pd.read_excel('mapped_output.xlsx', sheet_name='Details Sheet')
porsa_details 
```

```python
porsa_details.columns
```

```python
porsa_protection = porsa_details[['ROUTE NAME', 'ROAD STRUTURE TYPE', 'LENGTH (IN Mtr.)', 'PROTECTION TYPE','PROTECTION FOR', 'PROTECTION LENGTH (IN Mtr.)']]
```

```python
r_agg = porsa_protection.groupby(['ROUTE NAME', 'ROAD STRUTURE TYPE'])['LENGTH (IN Mtr.)'].agg(['count', 'sum']).unstack(fill_value=0)
r_agg
r_agg.columns = ['NO OF BRIDGE','NO OF CANAL' ,'NO OF CULVERT', 'LENGTH (IN Mtr) OF BRIDGE', 'LENGTH (IN Mtr) OF CANAL','LENGTH (IN Mtr) OF CULVERT']  # Rename columns
r_agg
r_agg = r_agg.reset_index()

# Aggregate sums for P (columns D, G)
p_agg = porsa_protection.pivot_table(values='PROTECTION LENGTH (IN Mtr.)', index='ROUTE NAME', columns='PROTECTION TYPE', aggfunc='sum', fill_value=0).reset_index()

# Merge the two aggregations
protection_ = pd.merge(r_agg, p_agg, on= 'ROUTE NAME', how='outer')
protection_
# Ensure column order matches df2
```

```python
merged_df.rename(columns={'span_name' : 'ROUTE NAME'} ,inplace=True)
protection_ = pd.merge(protection_, merged_df, on='ROUTE NAME', how='inner')
```
